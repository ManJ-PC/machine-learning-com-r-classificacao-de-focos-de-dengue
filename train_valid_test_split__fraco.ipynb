```python
# imports
import pandas as pd
from fast_ml.model_development import train_valid_test_split
```

```python
# load data
df = pd.read_csv('https://raw.githubusercontent.com/utkarshkant/Datasets-Helper-Functions/main/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
# preview data
df.head()
```

```python
# train, test, validation split
X_train, y_train, X_valid, y_valid, X_test, y_test = train_valid_test_split(df,
                                                                            target = 'Churn',
                                                                            train_size=0.6,
                                                                            valid_size=0.2,
                                                                            test_size=0.2)

# reset index
for data in [X_train, y_train, X_valid, y_valid, X_test, y_test]:
    data.reset_index(drop=True, inplace=True)
```

- The split datasets retain their original index and resetting it is an optional step.

```python
# split dataset size
print(X_train.shape, y_train.shape)
print(X_valid.shape, y_valid.shape)
print(X_test.shape, y_test.shape)
```
